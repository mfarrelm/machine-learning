```python
#Muhammad Farrel Mahendra
#Implementasi GA
class Population:
    def __init__(self, mutationRate, target, popSize,maxGen,chromClass):
        #create a population
        self.mutationRate = mutationRate
        self.target = target
        self.totalScore = 0
        self.popSize = popSize
        #self.maxGen = maxGen
        self.chromSet = []
        self.chromClass = chromClass 
        self.targetScore = len(target)**len(target)
        self.maxGen = maxGen
        self.generation = 0
            
    def createPopulation(self):
        #fill the population with chromosome
        for i in range (self.popSize):
            self.chromSet.append(self.chromClass(self.createRandomString()))
        
        
    def createRandomString(self):
        return ''.join(random.choice (string.ascii_lowercase+string.ascii_uppercase+' ') for i in range (len(self.target)))
    
    
    def pickParent(self): 
        rand1 = random.random()        
        rand2 = random.random()
       
        #index 1 and 2 is the index position of the parent in chromSet
        index1 = 0
        for chrom in self.chromSet:
            rand1 = rand1 - chrom.prob
            if (rand1 <=0 ) :
                break
            index1 = index1 + 1
        
        index2 = 0
        for chrom in self.chromSet:
            rand2 = rand2 - chrom.prob            
            if (rand2 <=0 ) :
                break
            index2 = index2 + 1
            
        return self.chromSet[index1], self.chromSet[index2]
        
    def calFitnessScore(self):
        self.totalScore = 0
        for chrom in self.chromSet :
        
            count = 0
            for index in range (len(self.target)):
                if chrom.value[index] == self.target[index]:
                    count += 1                
            chrom.score = count**count
            self.totalScore += chrom.score
        
        
    def calProb(self) :
        for chrom in self.chromSet:
            if (self.totalScore == 0) :
                chrom.prob = 0
            else :
                chrom.prob = chrom.score/self.totalScore
        
    
    def crossover(self,parent1,parent2):
        rand = random.random()
        newValue = parent1.value[:int(len(self.target)/2)] + parent2.value[int(len(self.target)/2):]
        child = self.chromClass(newValue)
        return child
        
    
    def mutation(self,child):
        for index in range (len(self.target)):
            char = random.choice (string.ascii_lowercase+string.ascii_uppercase+ ' ')
            prob = random.random()
        
            if(prob - self.mutationRate <= 0):
                child.value = child.value[:index] + char + child.value[index+1:]


            else:
                pass
            
    def printAllValue(self):
        for chrom in self.chromSet:
            print(chrom.value)
    
    def performAll(self):
        
        
        solutionFinded = False
        while (True) :    
            self.createPopulation()
            self.calFitnessScore()
            self.calProb()

            for index in range (self.popSize):
                if (Pop.chromSet[index].score >= self.targetScore):
                    solutionFinded = True
                    bestChromIndex = index
                    break

            if(solutionFinded):
                print(index)
                break

            childSet = []
            for i in range (self.popSize):
                parent1,parent2 = self.pickParent()
                child = self.crossover(parent1, parent2)
                self.mutation(child)
                childSet.append(child)
            self.chromSet = childSet
            print("generasi ke : %d     nilai : %f" %(self.generation,self.totalScore/self.popSize/self.targetScore))
            
            self.generation +=1
            if (self.generation >= self.maxGen):
                break

        

        
    
    
class Chromosome():
    def __init__(self, value):
        self.score = 0
        self.prob = 0
        self.value = value
    
    
```
